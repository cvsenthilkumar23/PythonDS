```python
mdata['CrossTraining']=mdata['CrossTraining'].fillna('NA')
```

```python
mdata['Category']=mdata['Category'].fillna('NA')
```

```python
mdata.isna().sum()
```

```python
mdata['Wall21'] = mdata['Wall21'].replace(' -   ', 0)
```

```python
mdata['Wall21'] = mdata['Wall21'].apply(pd.to_numeric, errors='ignore')
```

```python
marathon_data = mdata.drop(['CATEGORY', 'Category', 'CrossTraining', 'Marathon', 'Name', 'id'], axis=1)
```

```python
marathon_data.head()
```

```python
X= marathon_data.drop(['MarathonTime'],axis=1)
Y= marathon_data['MarathonTime']
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
print("Training feature set size:",X_train.shape)
print("Test feature set size:",X_test.shape)
print("Training variable set size:",y_train.shape)
print("Test variable set size:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
lm = LinearRegression() 
```

```python
lm.fit(X_train,y_train) 
```

```python
print("The intercept term of the linear model:", lm.intercept_)
```

```python
lm.score(X_test,y_test)
```

```python
train_pred = lm.predict(X_train)
```

```python
metrics.mean_squared_error(y_train,train_pred)
```

```python
metrics.mean_absolute_error(y_train,train_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_train,train_pred))
```

```python
import numpy as np

def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(y_train,train_pred)
```

```python
print("Mean absolute error (MAE):", metrics.mean_absolute_error(y_train,train_pred))
print("Mean square error (MSE):", metrics.mean_squared_error(y_train,train_pred))
print("Root mean square error (RMSE):", np.sqrt(metrics.mean_squared_error(y_train,train_pred)))
```

```python
print("R-squared value of this fit:",round(metrics.r2_score(y_train,train_pred),3))
```

```python
predictions = lm.predict(X_test)
```

```python
print("Mean absolute error (MAE):", metrics.mean_absolute_error(y_test,predictions))
print("Mean square error (MSE):", metrics.mean_squared_error(y_test,predictions))
print("Root mean square error (RMSE):", np.sqrt(metrics.mean_squared_error(y_test,predictions)))
```

```python
print("R-squared value of predictions:",round(metrics.r2_score(y_test,predictions),3))
```
